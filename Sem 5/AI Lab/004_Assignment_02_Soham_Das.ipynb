```python
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt

# Load Binary Dataset
binary_csv_path = "soham_binary.csv"
binary_df = pd.read_csv(binary_csv_path, header=None)

# Convert DataFrame to adjacency matrix
binary_adj_matrix = binary_df.values

# Create graph from adjacency matrix
G_binary = nx.from_numpy_array(binary_adj_matrix)

# Draw graph
plt.figure(figsize=(6, 6))
nx.draw(G_binary, with_labels=True, node_color='lightgreen', node_size=800, font_size=10)
plt.title("Graph from Binary Adjacency Matrix")
plt.show()
print(binary_adj_matrix)
```

```python
# Load Numerical Dataset
numerical_csv_path = "soham_numeric.csv"
numerical_df = pd.read_csv(numerical_csv_path, header=None)

# Convert DataFrame to adjacency matrix
numerical_adj_matrix = numerical_df.values

# Create weighted graph
G_numerical = nx.from_numpy_array(numerical_adj_matrix)

# Draw weighted graph with labels
plt.figure(figsize=(6, 6))
pos = nx.spring_layout(G_numerical)
nx.draw(G_numerical, pos, with_labels=True, node_color='lightgreen', node_size=800, font_size=10)

# Draw edge labels (weights)
edge_labels = nx.get_edge_attributes(G_numerical, 'weight')
nx.draw_networkx_edge_labels(G_numerical, pos, edge_labels=edge_labels)

plt.title("Weighted Graph from Numerical Adjacency Matrix")
plt.show()
```
